```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('results.csv')

# Extract the graph names and algorithm weights
graph_names = ['G' + str(name) for name in df['Name']]
algorithms = ['Rand', 'Greedy', 'SemiGreedy', 'LocalAvg', 'GRASP']
weights = df[algorithms].values.T  # Transpose to get algorithms as rows

# Set up the plot
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("husl")

# Create positions for the bars
x = range(len(graph_names))
bar_width = 0.15

# Plot each algorithm's results
for i, (algo, weight_values) in enumerate(zip(algorithms, weights)):
    plt.bar([pos + i * bar_width for pos in x], weight_values, 
            width=bar_width, label=algo)

# Customize the plot
plt.xlabel('Graph Instances')
plt.ylabel('Cut Weight')
plt.title('Comparison of Max-Cut Algorithms')
plt.xticks([pos + 2 * bar_width for pos in x], graph_names, rotation=45)
plt.legend(title='Algorithms')

# Adjust layout and save/show
plt.tight_layout()
plt.savefig('algorithm_comparison.png', dpi=300)
plt.show()
```
